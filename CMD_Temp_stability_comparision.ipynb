```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 12000)
T_NoiseASDn=0.1*((1+(3E-4/freq)**4))**0.5
T_NoiseASDo=0.5*((1+(2.828E-3/freq)**4))**0.5
plt.style.use('bmh')
# plot lines 
plt.plot(freq, T_NoiseASDn,'r-', label = "New") 
plt.plot(freq, T_NoiseASDo,'b-', label = "Old") 
plt.yscale("log")
plt.xscale("log")
plt.xlabel("Frequency (Hz)")
plt.ylabel("ASD(K/sqrt(Hz)") 
# Add a title to a legend 
plt.legend(title = "CMD Temp Stability") 
plt.show()
```
