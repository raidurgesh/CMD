```python
import numpy as np
import matplotlib.pyplot as plt
freq = np.linspace(0.00001, 1, 12000)
T_NoiseASD=0.1*((1+(3E-4/freq)**4))**0.5
plt.style.use('bmh')
plt.plot(freq, T_NoiseASD, 'r-',label = "ASD")
plt.yscale("log")
plt.xscale("log")
#plt.title("Temperature Instability ASD")
plt.xlabel("Frequency (Hz)")
plt.ylabel("ASD (K/sqrt(Hz)")
plt.legend(title = "Temp Instability")
plt.show()
```
